## Feature Engineering

```python
data_f.isna().sum() #no more missing values
```

## Qestions and Analysis

### Analyze trends in loan amounts over time for Mozambique

```python
import matplotlib.dates as mdates
#Convert end of period date 
data_f['End of Period'] = pd.to_datetime(data_f['End of Period'])

# Group the data by 'End of Period' and calculate total loan amounts for each period
loan_trends = data_f.groupby('End of Period')['Original Principal Amount'].sum().reset_index()

# Plot loan trends over time
plt.figure(figsize=(10, 6))
ax = loan_trends.plot('End of Period', 'Original Principal Amount', marker='o', linestyle='-')
ax.yaxis.set_major_formatter('{:,.0f}'.format)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%b'))
ax.set_ylim(bottom=0)
plt.title('Loan Trends for Mozambique')
plt.xlabel('End of Period')
plt.ylabel('Total Original Principal Amount')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
